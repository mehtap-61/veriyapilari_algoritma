#En uzun ismi ve bu ismin listenin içindeki indeksini bulan iki ayrı Phyton döngüsü yazma

```python
# İsim listesi
isim_listesi = ["Ali", "Şerafettin", "Osman", "İclal", "Ayşe"]

# En uzun ismi bulan fonksiyon
def en_uzun_ismi_bul(isim_listesi):
    en_uzun_isim = ""
    for isim in isim_listesi:
        if len(isim) > len(en_uzun_isim):
            en_uzun_isim = isim
    return en_uzun_isim

# En uzun ismin indeksini bulan fonksiyon
def en_uzun_isim_indeksi(isim_listesi):
    en_uzun_isim = en_uzun_ismi_bul(isim_listesi)
    en_uzun_isim_index = isim_listesi.index(en_uzun_isim)
    return en_uzun_isim_index

# En uzun ismi ve indeksini bulma
en_uzun_isim = en_uzun_ismi_bul(isim_listesi)
en_uzun_isim_index = en_uzun_isim_indeksi(isim_listesi)

# Sonuçları yazdırma
print("En uzun isim:", en_uzun_isim)
print("En uzun ismin indeksi:", en_uzun_isim_index)
```

1 ile 175 arasında 13'e tam bölüneleri bulan kodu yazın.

```python
Sayi_dizisi = []
sayi = 1

while sayi <= 175:
    if sayi % 13 == 0:
        Sayi_dizisi.append(sayi)
    sayi += 1

# Sonucu ekrana yazdır
print("1 ile 100 arasındaki 13'ye bölünen sayılar:", Sayi_dizisi)
print("Toplam sayı adeti: ", len(Sayi_dizisi))
```
